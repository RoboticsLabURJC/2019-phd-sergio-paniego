```python
import sys
import json
import yaml
import rosbag


bags = []
bag_1 = '20201228-145515.bag'
bag_2 = '20201228-145852.bag'
bag_3 = '20201228-150228.bag'
bag_4 = '20201228-150604.bag'
bag_5 = '20201228-150637.bag'
bag_6 = '20201228-150700.bag'
bag_7 = '20201228-150723.bag'
bag_8 = '20201228-151103.bag'

bags.append(bag_1)
bags.append(bag_2)
bags.append(bag_3)
bags.append(bag_4)
bags.append(bag_5)
bags.append(bag_6)
bags.append(bag_7)
bags.append(bag_8)


bags_checkpoints = []
bags_metadata = []

for bag_file in bags:
    print('------------------BAG----------------')
    bag = rosbag.Bag(bag_file)
    print(bag_file)
    bag_checkpoints = []
    for topic, point, t in bag.read_messages(topics=['/F1ROS/odom']):
        bag_checkpoints.append(point)
    bags_checkpoints.append(bag_checkpoints)
    
    for topic, point, t in bag.read_messages(topics=['/metadata']):
    
        y = yaml.load(str(point), Loader=yaml.FullLoader)
        h = json.dumps(y,indent=4)
        data = json.loads(h)
        metadata = json.loads(data['data'])
        #print(metadata)
        print(metadata['world'])
        print(metadata['brain_path'])
        print(metadata['robot_type'])
        bags_metadata.append(metadata)
        print(len(bag_checkpoints))
        
        
    bag.close()
    print('----------------------------------')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import metrics

experiments_statistics = []
world_completed = {}

for x, checkpoints in enumerate(bags_checkpoints):
    x_points = []
    y_points = []
    print('-------------------------------------------')
    print('WORLD -> ' + bags_metadata[x]['world'])
    print('BRAIN PATH -> ' + bags_metadata[x]['brain_path'])
    print('ROBOT TYPE -> ' + bags_metadata[x]['robot_type'])
    experiment_statistics = {}
    experiment_statistics['world'] = bags_metadata[x]['world']
    experiment_statistics['brain_path'] = bags_metadata[x]['brain_path']
    experiment_statistics['robot_type'] = bags_metadata[x]['robot_type']
    if bags_metadata[x]['world'] == 'simple_circuit.launch':
        perfect_lap_path = 'lap-simple-circuit.bag'
    elif bags_metadata[x]['world'] == 'many_curves.launch':
        perfect_lap_path = 'lap-many-curves.bag'
        
    perfect_lap_checkpoints, circuit_diameter = metrics.read_perfect_lap_rosbag(perfect_lap_path)
    print('CIRCUIT DIAMETER -> ' + str(circuit_diameter))
    lap_statistics = metrics.lap_percentage_completed(bags[x], perfect_lap_checkpoints, circuit_diameter)
    experiment_statistics['lap_statistics'] = lap_statistics
    experiments_statistics.append(experiment_statistics)
    print('COMPLETED DISTANCE -> ' + str(lap_statistics['completed_distance']))
    print('PERCENTAGE COMPLETED -> ' + str(lap_statistics['percentage_completed']))
    if lap_statistics['percentage_completed'] > 100:
        if bags_metadata[x]['world'] in world_completed and bags_metadata[x]['brain_path'] in world_completed[bags_metadata[x]['world']]:
            print('--- ENTRA 1 ---')
            world_completed[bags_metadata[x]['world']][bags_metadata[x]['brain_path']] = world_completed[bags_metadata[x]['world']][bags_metadata[x]['brain_path']] + 1
        elif bags_metadata[x]['world'] in world_completed:
            print('--- ENTRA 2 ---')
            world_completed[bags_metadata[x]['world']][bags_metadata[x]['brain_path']] = 1
        else:
            print('--- ENTRA 3 ---')
            world_completed[bags_metadata[x]['world']] = {} 
            world_completed[bags_metadata[x]['world']][bags_metadata[x]['brain_path']] = 1
            # world_completed[bags_metadata[x]['world']][bags_metadata[x]['brain_path']] = 1
    print('ORIENTATION MAE -> ' + str(lap_statistics['orientation_mae']))
    print('ORIENTATION TOTAL ERROR -> ' + str(lap_statistics['orientation_total_err']))
    
    if 'lap_seconds' in lap_statistics:
        print('LAP SECONDS -> ' + str(lap_statistics['lap_seconds']))
        print('CIRCUIT DIAMETER -> ' + str(lap_statistics['circuit_diameter']))
        print('AVERAGE SPEED -> ' + str(lap_statistics['average_speed']))

    for point in checkpoints:
        point_yml = yaml.load(str(point), Loader=yaml.FullLoader)
        # print(point)
        x_points.append(point_yml['pose']['pose']['position']['x'])
        y_points.append(point_yml['pose']['pose']['position']['y'])


    plt.scatter(x_points, y_points)
    plt.show()
    print('-------------------------------------------')
    
print(world_completed) 
print(experiments_statistics)
```

```python
import matplotlib.patches as mpatches

completed_laps = 0
experiment_names = []
experiments_completed_distances = []
experiments_percentage_completed = []
experiments_orientation_mae = []
experiments_orientation_total_err = []
experiments_lap_seconds = []
experiments_circuit_diameter = []
experiments_average_speed = []
circuit_bar_colors = []
for experimment_counter, experiment in enumerate(experiments_statistics):
    experiment_names.append(str(experimment_counter) + '. ' + experiment['brain_path'])
    experiments_completed_distances.append(experiment['lap_statistics']['completed_distance'])
    experiments_percentage_completed.append(experiment['lap_statistics']['percentage_completed'])
    experiments_orientation_mae.append(experiment['lap_statistics']['orientation_mae'])
    experiments_orientation_total_err.append(experiment['lap_statistics']['orientation_total_err'])
    if experiment['lap_statistics']['percentage_completed'] > 100:
        completed_laps += 1
        
    if 'lap_seconds' in experiment['lap_statistics']:
        experiments_lap_seconds.append(experiment['lap_statistics']['lap_seconds'])
        experiments_circuit_diameter.append(experiment['lap_statistics']['circuit_diameter'])
        experiments_average_speed.append(experiment['lap_statistics']['average_speed'])
    else:
        experiments_lap_seconds.append(0)
        experiments_circuit_diameter.append(0)
        experiments_average_speed.append(0)
        
    if experiment['world'] == 'simple_circuit.launch':
        circuit_bar_colors.append('black')
    else:
        circuit_bar_colors.append('red')

world_completed_circuit_colors = []
world_completed_brains = []
world_completed_brain_number = []
for key, value in world_completed.items():
    if key == 'simple_circuit.launch':
        world_completed_circuit_colors.append('black')
    else:
        world_completed_circuit_colors.append('red')

    for key_2, value_2 in value.items():
        world_completed_brains.append(key_2)
        world_completed_brain_number.append(value_2)
    

plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(experiment_names, experiments_completed_distances, color=circuit_bar_colors)
plt.suptitle('Completed distance for each brain', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('Completed distance', fontsize=60)
black_patch = mpatches.Patch(color='black', label='simple_circuit')
red_patch = mpatches.Patch(color='red', label='many_curves')
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()


plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(experiment_names, experiments_percentage_completed, color=circuit_bar_colors)
plt.suptitle('Completed percentage for each brain', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('Completed percentage', fontsize=60)
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()

plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(experiment_names, experiments_orientation_mae, color=circuit_bar_colors)
plt.suptitle('Orientation mae for each brain', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('Orientation mae', fontsize=60)
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()

plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(experiment_names, experiments_orientation_total_err, color=circuit_bar_colors)
plt.suptitle('Orientation total error for each brain', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('Orientation total error', fontsize=60)
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()

plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(experiment_names, experiments_lap_seconds, color=circuit_bar_colors)
plt.suptitle('Lap seconds for each brain', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('seconds', fontsize=60)
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()

plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(experiment_names, experiments_circuit_diameter, color=circuit_bar_colors)
plt.suptitle('Circuit diameter for each brain', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('meters', fontsize=60)
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()

plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(experiment_names, experiments_average_speed, color=circuit_bar_colors)
plt.suptitle('Average speed for each brain', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('speed', fontsize=60)
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()

plt.figure(figsize=(120, 20))
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=50)
plt.bar(world_completed_brains, world_completed_brain_number, color=world_completed_circuit_colors)
plt.suptitle('Completed laps for each circuit', fontsize=100)
plt.xlabel('Brain', fontsize=60)
plt.ylabel('Completed laps', fontsize=60)
plt.legend(handles=[black_patch, red_patch], loc=2, prop={'size': 60})
plt.show()
```
